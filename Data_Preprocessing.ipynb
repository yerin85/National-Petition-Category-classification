```
#파일읽어오기
import pandas as pd

data = pd.read_csv('/content/gdrive/My Drive/train_dataset.csv')
```

```
#결측값 제거
data = data.dropna(axis=0)

#카테고리가 기타인 값 제거
index_num = data[data['category']=='기타'].index
data = data.drop(index_num)
```

```
#카테고리 숫자화
data.loc[data["category"] == "정치개혁","category"] = 0
data.loc[data["category"] == "외교/통일/국방","category"] = 1
data.loc[data["category"] == "일자리","category"] = 2
data.loc[data["category"] == "미래","category"] = 3
data.loc[data["category"] == "성장동력","category"] = 4
data.loc[data["category"] == "농산어촌","category"] = 5
data.loc[data["category"] == "보건복지","category"] = 6
data.loc[data["category"] == "육아/교육","category"] = 7
data.loc[data["category"] == "안전/환경","category"] = 8
data.loc[data["category"] == "저출산/고령화대책","category"] = 9
data.loc[data["category"] == "행정","category"] = 10
data.loc[data["category"] == "반려동물","category"] = 11
data.loc[data["category"] == "교통/건축/국토","category"] = 12
data.loc[data["category"] == "경제민주화","category"] = 13
data.loc[data["category"] == "인권/성평등","category"] = 14
data.loc[data["category"] == "문화/예술/체육/언론","category"] = 15
```

```
import re

# 자음 ㄱ~ㅎ / 모음 ㅏ~ㅣ/ 가~힣 / 공백을 제외하고 모두 제거
train_content = data['content'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```
# ##train전처리파일 불러오는 코드
# X_train = pd.read_csv('/content/gdrive/My Drive/TLO/pre.csv')
```

```
from tensorflow.keras.utils import to_categorical

Y_train = to_categorical(data['category'])
```

```
#형태별로 나누어주기
from konlpy.tag import Okt
import tqdm
okt = Okt()

stopwords = ['있다','하다','을','를','이','가','의','은','들','좀','는','과','와','한','에','으로']
```

```
from tqdm import tqdm
#데이터 토큰화 (시간 좀 걸러요)
X_train=[]
for sentence in tqdm(train_content) : 
    temp_X = []
    temp_X = okt.morphs(sentence, stem = True)
    temp_X = [word for word in temp_X if not word in stopwords] #불용어제거
    temp_X = [word for word in temp_X if len(word) > 1 ] #토큰화하여 길이가 2 이상인 단어만 남기기
    X_train.append(temp_X)


import csv
with open('/content/gdrive/My Drive/smart_pre.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(X_train)
```

```
###########################Test SET 전처리 ################################
###########################################################################
test_data = pd.read_excel('/content/gdrive/My Drive/TLO/webcrawling_data.xlsx')
```

```
test_data.loc[test_data["category"] == "정치개혁","category"] = 0
test_data.loc[test_data["category"] == "외교/통일/국방","category"] = 1
test_data.loc[test_data["category"] == "일자리","category"] = 2
test_data.loc[test_data["category"] == "미래","category"] = 3
test_data.loc[test_data["category"] == "성장동력","category"] = 4
test_data.loc[test_data["category"] == "농산어촌","category"] = 5
test_data.loc[test_data["category"] == "보건복지","category"] = 6
test_data.loc[test_data["category"] == "육아/교육","category"] = 7
test_data.loc[test_data["category"] == "안전/환경","category"] = 8
test_data.loc[test_data["category"] == "저출산/고령화대책","category"] = 9
test_data.loc[test_data["category"] == "행정","category"] = 10
test_data.loc[test_data["category"] == "반려동물","category"] = 11
test_data.loc[test_data["category"] == "교통/건축/국토","category"] = 12
test_data.loc[test_data["category"] == "경제민주화","category"] = 13
test_data.loc[test_data["category"] == "인권/성평등","category"] = 14
test_data.loc[test_data["category"] == "문화/예술/체육/언론","category"] = 15
test_data.loc[test_data["category"] == "기타","category"] = 16
```

```
Y_test = test_data['category']
```

```
#데이터 인덱싱
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
```

```
threshold = 2
total_cnt = len(tokenizer.word_index) #단어수
rare_cnt = 0 #등장빈도수가 threshold보다 작은단어개수를 카운트
total_freq =0 #훈련데이터의 전체단어 빈도수 총합
rare_freq =0 #등장빈도수가 threshold보다 작은단어의 등장빈도수 총합

for key, value in tokenizer.word_counts.items():
    total_freq = total_freq + value

    if(value < threshold) : 
        rare_cnt = rare_cnt +1
        rare_freq = rare_freq + value

print('단어 집합의 크기: ', total_cnt)
print('등장빈도가 %s번 이하인 희귀단어의 수 : %s'%(threshold-1, rare_cnt))
print("단어 집합에서 희귀단어의 비율: ", (rare_cnt/total_cnt)*100)
print("전체 등장빈도에서 희귀단어등장 빈도 비율 : ", (rare_freq/total_freq)*100)
```

```
vocab_size = total_cnt - rare_cnt + 1 #전체단어개수중 빈도수가 2이하인 단어는 제거
print("단어집합의 크기 : ", vocab_size)
```

```
tokenizer = Tokenizer(vocab_size)
tokenizer.fit_on_texts(X_train)
#X_train만 fit하여 이를 기준으로 학습데이터와 테스트데이터를 인덱싱화
X_train = tokenizer.texts_to_sequences(X_train)
```

```
X_test = tokenizer.texts_to_sequences(X_test)
```

```
print('train data의 최대길이 : ' ,max(len(I) for I in X_train))
print('train data의 평균길이 : ', sum(map(len, X_train))/len(X_train))
print('test data의 최대길이 : ' ,max(len(I) for I in X_test))
print('test data의 평균길이 : ', sum(map(len, X_test))/len(X_test))
```

```
def below_threshold_len(max_len, nested_list):
    cnt=0
    for i in nested_list : 
        if(len(i) <= max_len) :
            cnt = cnt+1
    print("전체 샘플 중 길이가 %s 이하인 샘플의 비율 : %s" %(max_len, (cnt/ len(nested_list))*100))
```

```
max_len = 300
below_threshold_len(max_len, X_train)
```

```
import numpy as np
from keras.preprocessing import sequence
from keras.preprocessing.sequence import pad_sequences

#데이터 패딩
X_train = pad_sequences(X_train, maxlen = max_len)
X_test=pad_sequences(X_test, maxlen=max_len)
```

```
#모델생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.models import load_model

model = Sequential()
model.add(Embedding(vocab_size, 32, input_length=max_len)) 
#Embedding층은 전처리과정을 통해 입력된 값을 받아 다음층이 알아들을 수 있는 형태로 변환
#Embedding('불러온 단어의 개수','기사당 단어수')
model.add(LSTM(32))
#LSTM은 RNN에서 기억 값에 대한 가중치를 제어
#LSTM(기사당 단어수, 기타옵션)
model.add(Dense(17,  activation = 'softmax'))
model.summary()
```

```
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics=['acc'])
```

```
history = model.fit(X_train, Y_train, batch_size =128, epochs =3)
```

```
y_pred = model.predict_classes(X_test)
```

```
cnt = 0
for i in Y_test : 
    if(Y_test[i] == y_pred[i]) :
        cnt = cnt + 1

print("실제 카테고리 분류와 예측한 카테고리의 일치율 :" ,(cnt/len(Y_test))*100)
```
